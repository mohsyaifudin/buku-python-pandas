```python
import pandas as pd
import matplotlib.pyplot as plt

# Data penjualan per bulan
data = {'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Penjualan': [1000, 1500, 2000, 1800, 2200]}

df = pd.DataFrame(data)

# Membuat grafik garis untuk menunjukkan tren penjualan
plt.plot(df['Bulan'], df['Penjualan'], marker='o')
plt.title('Tren Penjualan Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.grid(True)
plt.show()
```

```python
# Data penjualan per kategori produk
data_kategori = {'Kategori': ['Elektronik', 'Furnitur', 'Pakaian'],
                 'Jan': [5000, 3000, 2000],
                 'Feb': [6000, 3500, 2500],
                 'Mar': [7000, 4000, 3000]}

df_kategori = pd.DataFrame(data_kategori)

# Membuat grafik bar untuk membandingkan penjualan per kategori
df_kategori.set_index('Kategori').T.plot(kind='bar')
plt.title('Perbandingan Penjualan per Kategori')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.xticks(rotation=0)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Data penjualan per bulan
data = {'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Penjualan': [1000, 1500, 1800, 1700, 2000]}

df = pd.DataFrame(data)

# Membuat grafik garis
plt.plot(df['Bulan'], df['Penjualan'], marker='o', linestyle='-', color='b')
plt.title('Tren Penjualan Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')
plt.grid(True)
plt.show()
```

```python
# Menambahkan kustomisasi
plt.plot(df['Bulan'], df['Penjualan'], marker='s', linestyle='-', color='r')
plt.title('Tren Penjualan Bulanan', fontsize=14, color='green')
plt.xlabel('Bulan', fontsize=12, color='blue')
plt.ylabel('Penjualan', fontsize=12, color='blue')
plt.grid(True, linestyle='--', color='gray')
plt.xticks(rotation=45)  # Rotasi label bulan
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data tinggi badan siswa
data_tinggi = {'Tinggi': [150, 160, 165, 170, 180, 155, 145, 170, 185, 190]}

df_tinggi = pd.DataFrame(data_tinggi)

# Membuat boxplot menggunakan Seaborn
sns.boxplot(data=df_tinggi, x='Tinggi', color='lightblue')
plt.title('Distribusi Tinggi Badan Siswa')
plt.xlabel('Tinggi Badan (cm)')
plt.show()
```

```python
# Data harga dan penjualan
data_harga_penjualan = {'Harga': [100, 150, 200, 250, 300],
                        'Penjualan': [5000, 4500, 4000, 3500, 3000]}

df_harga_penjualan = pd.DataFrame(data_harga_penjualan)

# Membuat scatter plot dengan Seaborn
sns.scatterplot(data=df_harga_penjualan, x='Harga', y='Penjualan', color='red', marker='o')
plt.title('Hubungan antara Harga dan Penjualan')
plt.xlabel('Harga')
plt.ylabel('Penjualan')
plt.show()
```

```python
# Membuat boxplot dengan palet warna berbeda
sns.set_palette("pastel")  # Menggunakan palet pastel
sns.boxplot(data=df_tinggi, x='Tinggi')
plt.title('Distribusi Tinggi Badan Siswa dengan Palet Pastel')
plt.show()
```

```python
# Membuat scatter plot dengan garis regresi
sns.regplot(data=df_harga_penjualan, x='Harga', y='Penjualan', color='blue', line_kws={'color': 'green'})
plt.title('Hubungan antara Harga dan Penjualan dengan Regresi')
plt.xlabel('Harga')
plt.ylabel('Penjualan')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Data contoh
data = {'Var1': [1, 2, 3, 4, 5],
        'Var2': [2, 4, 6, 8, 10],
        'Var3': [5, 4, 3, 2, 1],
        'Var4': [1, 1, 1, 1, 1]}

df = pd.DataFrame(data)

# Menghitung korelasi antar variabel
corr_matrix = df.corr()

# Membuat heatmap untuk korelasi antar variabel
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap Korelasi Antar Variabel')
plt.show()
```

```python
# Data tinggi badan siswa
data_tinggi = {'Tinggi': [150, 160, 165, 170, 180, 155, 145, 170, 185, 190]}

df_tinggi = pd.DataFrame(data_tinggi)

# Membuat boxplot untuk memvisualisasikan distribusi tinggi badan
sns.boxplot(data=df_tinggi, x='Tinggi', color='lightgreen')
plt.title('Distribusi Tinggi Badan Siswa')
plt.xlabel('Tinggi Badan (cm)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data penjualan per bulan
data = {'Bulan': ['Jan', 'Feb', 'Mar', 'Apr', 'May'],
        'Penjualan': [1000, 1500, 2000, 1700, 2200]}

df = pd.DataFrame(data)

# Membuat grafik garis dengan warna khusus
plt.plot(df['Bulan'], df['Penjualan'], marker='o', linestyle='-', color='#FF5733')  # Warna dengan kode hex
plt.title('Tren Penjualan Bulanan', fontsize=14)
plt.xlabel('Bulan', fontsize=12)
plt.ylabel('Penjualan', fontsize=12)
plt.grid(True)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Data penjualan per kategori
data = {'Kategori': ['Elektronik', 'Furnitur', 'Pakaian'],
        'Penjualan': [1000, 1500, 1200]}

df = pd.DataFrame(data)

# Membuat grafik batang dengan palet warna 'pastel'
sns.barplot(x='Kategori', y='Penjualan', data=df, palette='pastel')
plt.title('Penjualan per Kategori', fontsize=14)
plt.xlabel('Kategori', fontsize=12)
plt.ylabel('Penjualan', fontsize=12)
plt.show()
```

```python
# Membuat palet warna kustom
custom_palette = sns.color_palette(["#3498db", "#e74c3c", "#2ecc71"])

# Membuat grafik batang dengan palet warna kustom
sns.barplot(x='Kategori', y='Penjualan', data=df, palette=custom_palette)
plt.title('Penjualan per Kategori', fontsize=14)
plt.show()
```

```python
# Membuat grafik garis dengan label dan judul yang disesuaikan
plt.plot(df['Bulan'], df['Penjualan'], marker='o', color='b')
plt.title('Tren Penjualan per Bulan', fontsize=14, color='green')
plt.xlabel('Bulan', fontsize=12, color='blue')
plt.ylabel('Penjualan', fontsize=12, color='blue')
plt.grid(True)
plt.show()
```

```python
# Membuat grafik batang dengan label dan judul yang disesuaikan
sns.barplot(x='Kategori', y='Penjualan', data=df, palette='muted')
plt.title('Penjualan per Kategori', fontsize=16, color='darkred')
plt.xlabel('Kategori Produk', fontsize=14)
plt.ylabel('Jumlah Penjualan', fontsize=14)
plt.show()
```

```python
# Menggunakan tema 'darkgrid' untuk grafik
sns.set_theme(style="darkgrid")

# Membuat grafik batang dengan tema yang disesuaikan
sns.barplot(x='Kategori', y='Penjualan', data=df, palette='deep')
plt.title('Penjualan per Kategori', fontsize=16, color='black')
plt.xlabel('Kategori', fontsize=14)
plt.ylabel('Penjualan', fontsize=14)
plt.show()
```

```python
# Menyesuaikan tema dengan palet warna kustom
sns.set_theme(style="whitegrid", palette="muted")

# Membuat grafik dengan tema kustom
sns.lineplot(x=df['Bulan'], y=df['Penjualan'], marker='o')
plt.title('Tren Penjualan Bulanan', fontsize=16, color='darkblue')
plt.xlabel('Bulan', fontsize=14)
plt.ylabel('Penjualan', fontsize=14)
plt.show()
```

```python
# Menyesuaikan tema dengan palet warna kustom
sns.set_theme(style="whitegrid", palette="muted")

# Membuat grafik dengan tema kustom
sns.lineplot(x=df['Bulan'], y=df['Penjualan'], marker='o')
plt.title('Tren Penjualan Bulanan', fontsize=16, color='darkblue')
plt.xlabel('Bulan', fontsize=14)
plt.ylabel('Penjualan', fontsize=14)
plt.show()
```

```python
# Membuat grafik garis dan menambahkan anotasi
plt.plot(df['Bulan'], df['Penjualan'], marker='o', color='purple')
plt.title('Tren Penjualan Bulanan')
plt.xlabel('Bulan')
plt.ylabel('Penjualan')

# Menambahkan anotasi pada titik tertentu
plt.annotate('Penurunan pada April', xy=('Apr', 1700), xytext=('Feb', 1800),
             arrowprops=dict(facecolor='red', arrowstyle='->'))

plt.show()
```

```python
# Membuat grafik batang dengan anotasi
sns.barplot(x='Kategori', y='Penjualan', data=df, palette='pastel')
plt.title('Penjualan per Kategori')

# Menambahkan anotasi pada batang tertentu
for index, value in enumerate(df['Penjualan']):
    plt.text(index, value + 50, f'{value}', ha='center', fontsize=12)

plt.show()
```
